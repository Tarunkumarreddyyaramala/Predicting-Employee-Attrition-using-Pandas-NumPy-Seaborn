# 3. Exploratory Data Analysis (EDA)

## 3.1 Education Distribution

```python
ax= df['Education'].value_counts().head(20).plot(kind='bar',title=' Distribution of Employees')
ax.set_xlabel('Educatiom')
ax.set_ylabel('Count')
plt.show()
```

#### Observation:

Bar chart shows the distribution of employees by education level.

## 3.2 Age Distribution

```python
ax = df['Age'].plot(kind = 'hist',bins = 20, title = 'employee ages concentrated between 25–40 years')
ax.set_xlabel('Age')
plt.show()
```

#### Observation:

Histogram shows employee ages concentrated between 25–40 years.

## 3.3 Performance Rating

```python
df['PerformanceRating'].plot(kind= 'kde')
plt.show()
```

#### Observation:

KDE plot shows performance ratings mostly skewed toward higher ratings.

## 3.4 EmployeeNumber vs HourlyRate

```python
df.plot(kind= 'scatter',title = ' EmpolyeeNumber vs HourlyRate' ,x='EmployeeNumber', y = 'HourlyRate')
plt.show()
```

#### Observation:

Scatter plot confirms that EmployeeNumber (ID) is not related to HourlyRate.

## 3.5 Age vs HourlyRate by JobRole

```python
sns.scatterplot(x='Age', y = 'HourlyRate',hue = 'JobRole',data =df)
plt.show()
```

#### Observation:

Scatter plot shows different clusters of hourly rates across job roles.

## 3.6 Pairplot

```python
sns.pairplot(df, vars = ['Age', 'Department', 'DistanceFromHome', 'Education','Gender',
       'HourlyRate', 'JobInvolvement', 'JobLevel', 'JobRole'])
plt.show()
```

#### Observation:

Pairplot visualizes relationships among selected variables.

Some weak correlations can be seen (e.g., JobLevel vs HourlyRate)

## 3.7 Correlation Matrix

```python
df_corr = df[['Age','DistanceFromHome', 'Education',
       'HourlyRate', 'JobInvolvement', 'JobLevel']].dropna().corr()
df_corr
```

#### Observation:

Correlation matrix shows numeric relationships.

## 3.8 Correlation Heatmap

```python
sns.heatmap(df_corr,annot= True)
plt.show()
```

#### Observation:

Heatmap highlights a strong correlation between JobLevel and HourlyRate.

Other variables show weaker or negligible correlation.
